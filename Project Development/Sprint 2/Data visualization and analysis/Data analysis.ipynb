```python
import numpy as np
import seaborn as sb 
import pandas as pd
from pandas_profiling import ProfileReport
import plotly.express as px
import plotly.graph_objects as go
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("loan_prediction.csv")
df.head()
```

```python
df.info()
```

```python
profile = ProfileReport(df, title="Analysis report for data Analysis")
profile.to_notebook_iframe()
profile.to_file("data_analysis.html")
```

## Univariate Analysis 
- Univariate analysis is basically the simplest form to analyze data. Uni means one and this means that the data has only one kind of variable

```python
sb.displot(data=df, x="ApplicantIncome",kde=True)
```

```python
fig = px.histogram(df, x="Education",color="Education")
fig.show()
```

```python
fig = px.histogram(df, x="Gender",color="Gender")
fig.show()
```

```python
fig = px.histogram(df, x="Loan_Status",color="Loan_Status")
fig.show()
```

```python
fig = go.Figure(data=[go.Histogram(x=df["LoanAmount"],marker_color='#80489C')])
fig.show()
```

```python
sb.barplot(x='ApplicantIncome', y='LoanAmount',data=df)
plt.show()
```

```python
fig = px.scatter(df, x="LoanAmount", y="Loan_Amount_Term")
fig.show()
```

```python
fig = px.strip(df, x="Gender", y="ApplicantIncome",color="Gender")
fig.show()
```

```python
fig = px.strip(df, x="Education", y="ApplicantIncome",color="Education")
fig.show()
```

```python
fig = px.scatter_matrix(df)
fig.update_layout(width=1500,height=2000)
fig.show()
```
